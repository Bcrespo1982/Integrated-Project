````python
    # Primeira hipótese (Plataformas Xbox One e PC):

    # Hipótese nula (H₀): As classificações médias dos usuários nas plataformas Xbox One e PC são as mesmas.

    # Hipótese alternativa (H₁): As classificações médias dos usuários nas plataformas Xbox One e PC são diferentes.

    # Segunda hipótese: Hipótese 2: Comparação entre gêneros Action e Sports:

    # Hipótese nula (H₀): As classificações médias dos gêneros Action e Sports são iguais.
    
    # H₀: μAction=μSports
    
    # Hipótese alternativa (H₁):
    
    # As classificações médias dos gêneros Action e Sports são diferentes.
    
    # H₁: μAction≠μSports
````

````python
from scipy import stats

# Filtrando os dados para as plataformas Xbox One e PC
x360_scores = games[games['platform'] == 'x360']['user_score'].dropna()
pc_scores = games[games['platform'] == 'pc']['user_score'].dropna()

# Teste t para as plataformas Xbox One e PC
t_stat_1, p_value_1 = stats.ttest_ind(x360_scores, pc_scores)

# Exibindo os resultados
print(f"Teste para Xbox One vs PC: T-Stat = {t_stat_1}, P-Value = {p_value_1}")

````

````python
# Filtrando os dados para os gêneros Action e Sports
action_scores = games[games['genre'] == 'action']['user_score'].dropna()
sports_scores = games[games['genre'] == 'sports']['user_score'].dropna()

# Teste t para os gêneros Action e Sports
t_stat_2, p_value_2 = stats.ttest_ind(action_scores, sports_scores)

# Exibindo os resultados
print(f"Teste para Action vs Sports: T-Stat = {t_stat_2}, P-Value = {p_value_2}")
````

````python
# Como esse valor-p é maior que o nível de significância padrão α = 0.05, não temos evidência suficiente para rejeitar a hipótese nula.
````
